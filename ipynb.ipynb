

This dataset contains weekly retail prices of three grades of gasoline, diesel, auto propane and compressed natural gas in 10 Ontario markets. It is downloaded from https://data.ontario.ca/dataset/fuels-price-survey-information. Please plot the following visualizations. 




```python
import pandas as pd
df = pd.read_csv('data/fueltypesall.csv')
```



Compare the mean Regular Unleaded Gasoline price among different cities in 2019. Plot the corresponding bar chart.

```python
import matplotlib.pyplot as plt 
df = pd.read_csv('data/fueltypesall.csv')
df = df[['Date','Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor',
       'London', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay',
       'Timmins', 'Fuel Type']] 
df = df[df["Fuel Type"] == "Regular Unleaded Gasoline"]
df = df[(df["Date"] > "2019-01-01") & (df["Date"] < "2019-12-31")]
d = {'Cities': ['Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor',
       'London', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay',
       'Timmins'], 
     'Mean Price': [df.loc[:,"Ottawa"].mean(),
                   df.loc[:,"Toronto West/Ouest"].mean(),
                   df.loc[:,"Toronto East/Est"].mean(),
                   df.loc[:,"Windsor"].mean(),
                   df.loc[:,"London"].mean(),
                   df.loc[:,"Sudbury"].mean(),
                   df.loc[:,"Sault Saint Marie"].mean(),
                   df.loc[:,"Thunder Bay"].mean(),
                   df.loc[:,"North Bay"].mean(),
                   df.loc[:,"Timmins"].mean()]
    }
df = pd.DataFrame(data=d)

fig, ax = plt.subplots(figsize=(15,5))
plt.bar(df["Cities"],df["Mean Price"],width=0.5, hatch="\\")
plt.xlabel('Cities', fontsize=15)
plt.ylabel('Mean Regular Unleaded Gasoline Price', fontsize=15)
plt.show()
```


Plot the density distribution (with Gaussian kernel) of the Regular Unleaded Gasoline price in Ottawa for 2019. Choose the appropriate parameters.

```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt 
df = pd.read_csv('data/fueltypesall.csv')
sns.set_style("white")
#import data

df = df[['Date','Ottawa', 'Fuel Type']]
df = df[df["Fuel Type"] == "Regular Unleaded Gasoline"]
df = df[(df["Date"] > "2019-01-01") & (df["Date"] < "2019-12-31")]
df[['Ottawa']].plot.kde(color='green') 
plt.title('Gaussian Kernel Density Distribution of Regular Unleaded Gasoline Price in Ottawa for 2019') 
plt.show()
```


Plot the boxplot for the Regular Unleaded Gasoline price in Ottawa across year 2012-2019. The distribution of each year should be visualized seperately.

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
df = pd.read_csv('data/fueltypesall.csv')
df = df[['Date','Ottawa', 'Fuel Type']] 
df = df[df["Fuel Type"] == "Regular Unleaded Gasoline"]
df = df[(df["Date"] > "2012-01-01") & (df["Date"] < "2019-12-31")]

df.set_index('Date',inplace=True)
d = {
    '2012': df.loc['2012-01-01':'2012-12-31', 'Ottawa'],
    '2013': df.loc['2013-01-01':'2013-12-31', 'Ottawa'],
    '2014': df.loc['2014-01-01':'2014-12-31', 'Ottawa'],
    '2015': df.loc['2015-01-01':'2015-12-31', 'Ottawa'],
    '2016': df.loc['2016-01-01':'2016-12-31', 'Ottawa'],
    '2017': df.loc['2017-01-01':'2017-12-31', 'Ottawa'],
    '2018': df.loc['2018-01-01':'2018-12-31', 'Ottawa'],
    '2019': df.loc['2019-01-01':'2019-12-31', 'Ottawa']
    }

fig, ax = plt.subplots()
ax.boxplot(d.values(), showmeans=True, notch=True)
ax.set_xticklabels(d.keys())
plt.show()
```
